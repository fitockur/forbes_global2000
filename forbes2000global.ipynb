```python
import time
from selenium import webdriver

driver = webdriver.Chrome(PATH_TO_WEBDRIVER)  # Optional argument, if not specified will search path.
driver.get('https://www.forbes.com/global2000/list/')

SCROLL_PAUSE_TIME = 1

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

content = driver.page_source
driver.quit()
```

```python
soup = BeautifulSoup(content)
```

```python
print(soup.prettify())
```

```python
dirty_list = soup.find('tbody').find_all('tr', {'class': 'data'})
```

```python
from tqdm import tqdm_notebook
```

```python
new_data = []
for url in tqdm_notebook(df.href):
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    rows = soup.find('div', {'class': "profile-info"}).find_all('div', {'class': "profile-row"})
    cur_dict = {}
    cur_row = []
    for row in rows:
        if row.find_all('span'):
            cur_dict[row.find_all('span')[0].text] = row.find_all('span')[1].text
    for column in new_columns:
        cur_row.append(cur_dict.get(column))
    new_data.append(cur_row)
```

```python
res = pd.concat([df[['Rank', 'Company', 'Country/Territory', 'Sales', 
                     'Profits', 'Assets', 'Market Value']], 
                 new_df[['Industry', 'Founded', 'Chief Executive Officer',
                         'Employees']]], axis=1).fillna(float('nan'))

print(res.shape)
res.head(10)
```

```python
res.info()
```

```python
def correct_money(money):
    try:
        _money = float(money[1:-1].replace(',', ''))
        if money[-1] == 'M':
            return _money * 1e6
        elif money[-1] == 'B':
            return _money * 1e9
        else:
            return money
    except ValueError:
        return float('nan')
    

def correct_empl(employees):
    if isinstance(employees, float):
        return float('nan')
    _empl = employees.replace(',', '')
    try:
        return int(_empl)
    except ValueError:
        return float('nan')
    
res['Sales'] = res['Sales'].apply(correct_money)
res['Profits'] = res['Profits'].apply(correct_money)
res['Assets'] = res['Assets'].apply(correct_money)
res['Market Value'] = res['Market Value'].apply(correct_money)
res['Employees'] = res['Employees'].apply(correct_empl)
```
